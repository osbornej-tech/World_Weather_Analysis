```python
import random
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import requests
```

```python
from config import weather_api_key
from citipy import citipy
```

```python
# Create a set of random latitude and longitude combinations.
lat = np.random.uniform(low=-90.000, high=90.000, size=2000)
lon = np.random.uniform(low=-180.000, high=180.000, size=2000)
lat_lngs = zip(lat, lon)
lat_lngs
```

```python
cities = []
for i in lat_lngs:
    city = citipy.nearest_city(i[0], i[1]).city_name
    if city not in cities:
        cities.append(city)
len(cities)
```

```python
#url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key
url = 'http://api.openweathermap.org/data/2.5/weather?'
units = 'imperial'
query_url = f"{url}appid={weather_api_key}&units={units}&q="
#query_url = url+'lat={'+ + '}&lon={}'+ str(lon)+'}&appid='+weather_api_key 
print(query_url)
```

```python
lat= []
lon = []
country = []
max_temp =[]
humidity =[]
cloudiness = []
wind_speed = []
description =[]
cities_name =[]

record_count = 1
set_count = 1

# Loop through all the cities in our list
for i, city in enumerate(cities):
    
#for i, city in enumerate(cities[:5]):   
#run first 5

    # Group cities in sets of 50 for logging purposes
    if (i % 50 == 0 and i >= 50):
        set_count += 1
        record_count = 0
       
    try:
        
        response = requests.get(query_url+city).json()
        country.append(response['sys']['country'])
        lat.append(response['coord']['lat'])
        lon.append(response['coord']['lon'])
        max_temp.append(response['main']['temp_max'])
        humidity.append(response['main']['humidity'])
        cloudiness.append(response['clouds']['all'])
        wind_speed.append(response['wind']['speed'])
        description.append(response['weather'][0]['description'])
        cities_name.append(city)
        print(f'City name of: {city}, {record_count}, {set_count}')
        record_count += 1
    except:
        print ("City not found")
```

```python
# Capitalize city names
cities_name = map(lambda x:x.capitalize(),cities_name)
```

```python
city_df = pd.DataFrame({'City': cities_name,'Country': country, 
                        'Lat': lat,'Lng': lon, 'Max Temp': max_temp, 
                        "Humidity": humidity, 'Cloudiness': cloudiness,
                        'Wind Speed': wind_speed,'Current Description': description})
city_df.head(10)
```

```python
# Save the DataFrame as a csv
city_df.to_csv("WeatherPy_Database.csv", encoding="utf-8", index=False)

#output_data_file = "WeatherPy_Database.csv"
#city_data_df.to_csv(output_data_file, index_label="City_ID")
```
